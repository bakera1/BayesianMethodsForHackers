```python
import matplotlib.pyplot as plt
from scipy.stats import poisson
import numpy as np
```

```python
fig = plt.figure(figsize=(12.5,4))
a = np.arange(16)
poi = poisson
colors = ['#34BABD', '#A60628']
lambda_ = [1.5, 4.2]
plt.bar(a, poi.pmf(a, lambda_[0]), color=colors[0], label="$ \lambda={0:.2f} $".format(lambda_[0]), alpha=0.60, edgecolor=colors[0], lw="3")
plt.bar(a, poi.pmf(a, lambda_[1]), color=colors[1], label="$ \lambda={0:.2f} $".format(lambda_[1]), alpha=0.60, edgecolor=colors[1], lw="3")
plt.grid(True)
plt.legend()
plt.ylabel("Probaiblity of $k$")
plt.xlabel("$k$")
plt.title("Probability mass function of a Poisson random variable, differing $\lambda$ values.")
```

```python
fig = plt.figure(figsize=(12.5,4))
a = np.linspace(0,4,100)
from scipy.stats import expon
lambda_ = [0.5,1]

for (l,c) in zip(lambda_, colors):
    plt.plot(a, expon.pdf(a, scale=1./l), lw=3, color=c, label="$ \lambda {0:.2f}".format(l))
    plt.fill_between(a, expon.pdf(a, scale=1./l), color=c, alpha=0.33)
    
plt.grid(True)
plt.ylabel("Probability of density funciton at $z$")
plt.xlabel("$z$")
plt.ylim(0, 1.2)
plt.title("Probability density function of an exponential random variable, differing $\lambda$ values")
```

```python
from scipy.stats import norm
nor = norm
x = np.linspace(-8, 7 , 150)
mu = (-2, 0, 3)
tau = (.7, 1, 2.8)
colors = ['#34BABD', '#A60628', '#7A68A6']
parameters = zip(mu, tau, colors)

fig = plt.figure(figsize=(12.5,6))
for _mu, _tau, _color in parameters:
    plt.plot(x, nor.pdf(x, _mu, scale=1./_tau), label="$ \mu={0:.1f}, \\tau={1:.1f} $".format(_mu, _tau), color=_color)
    plt.fill_between(x, nor.pdf(x, _mu, scale=1./_tau), color=_color, alpha=.33)
plt.legend(loc="upper right")
plt.grid(True)
plt.title("Probability distribution of three different Normal random variables.")
plt.xlabel("$x$")
plt.ylabel("Density funciton at $x$")
```
